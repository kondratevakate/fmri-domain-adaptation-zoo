```python
import seaborn as sns
# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

fmri.head()
```

```python
fmri_restricted = fmri[fmri['subject'].isin(['s1','s2', 's3', 's4', 's5'])]
fmri_restricted.rename(columns = {'subject': 'ROIs'}, inplace = True)
fmri_restricted['ROIs'] = 'ROI_' + fmri_restricted['ROIs'].str[1:]
```

```python
sns.set(rc={'figure.figsize':(5,4)})
sns.set_style("white")

# sns.set_context("paper")
sns.set_context("talk")
sns.axes_style("white")
sns_plot = sns.lineplot(x="timepoint", y="signal",
             hue="ROIs",
             data=fmri_restricted)
fig = sns_plot.get_figure()

fig.savefig("fmri_ts.png")
```

```python
sns.set_style("whitegrid")
sns.despine(left=True)
sns.set_context("talk")
sns_plot = sns.lineplot(x="timepoint", y="signal",
             hue="ROIs",
             data=fmri_restricted,  palette="deep")
```

```python
fig = sns_plot.get_figure()

fig.savefig("fmri_ts_2.png")
```
